plotting sin linea. para trazar solo los marcadores , puede usar notacion de cadena de acceso directo 'o', que significa "anillos",Ejemolo:Dibuja dos puntos en el diagrama, uno en la posicion (1,3) y otro en la posicion (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

punrtosx = np.array([1,8 ])
puntosy = np.array ([3,10 ])
plt.plot(puntosx, puntosy, 'o')
plt.show()
```

diagramas de dispercion (scatter)

se emplean para mostrar la relacion entre dos variables numericas, detectando correlacionales o agrupamientos
Ejemplo

```python
import matplotlib.pyplot as plt

altura = [150,160,170,180,190,200]
peso = [50,60,65,80,85,95]

plt.scatter(altura,peso,color = "red" , marker = "x")
plt.title("relacion entre la altura y el peso")
plt.xlabel("altura (cms)")
plt.ylabel("peso(kg)")
plt.show()
```

Graficos de barras (bar y barch)
sirven para comparar categorias o variables tanto en orientacion vartical como horizontal.
Ejemplo de barras verticales:

```python
import matplotlib.pyplot as plt

categorias = ["A","B","C","D" ]
valores = [4,7,1,8 ]

plt.bar(categorias,valores, color = "orange")
plt.title("grafica de barras ")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
